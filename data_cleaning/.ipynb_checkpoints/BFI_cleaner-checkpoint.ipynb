```python
import pandas as pd
from numpy import array
```

```python
raw_data = pd.read_csv("Personality (BFI).csv")
raw_data
```

```python
# Replace text responses with numeric responses
data = raw_data
data = data.replace({'Disagree Strongly':1}, regex=True)
data = data.replace({'Disagree a little':2}, regex=True)
data = data.replace({'Neither agree nor disagree':3}, regex=True)
data = data.replace({'Agree a little':4}, regex=True)
data = data.replace({'Agree Strongly':5}, regex=True)
data
```
